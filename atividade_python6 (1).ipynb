```python
# neste aqui eu concertei o erro de colocarem alguma coisa diferente de numeros no def trocar_B_A
# e se caso a pessoa colocar o valor 0 pedir par trocar msm que nao da erro no codigo, nao faz sentido o objeto ter tamanho 0
class Retagulo:
    def __init__(self,Base,Altura):
        self.Base = Base
        self.Altura = Altura
        
    def trocar_B_A(self):
        try:
            self.Base = float(input("qual valor da Base:  "))
            self.Altura = float(input("qual valor da Altura   "))
            print("Base: ",self.Base , "Altura: ",self.Altura)
       
        except ValueError:
            print("apenas numeros")
        
    
    def calc_area(self):
        if self.Base == 0 or self.Altura == 0:
            print("O valor da Base ou Altura nao pode ser 0")
            print("Troque o valor")
        else:
            print(self.Base * self.Altura)
        
    def calc_peri(self):
        if self.Base == 0 or self.Altura == 0:
            print("O valor da Base ou Altura nao pode ser 0")
            print("Troque o valor")
        else:
            print((self.Base + self.Altura)*2)
        
    def mostraLado(self):
        print("Base: ",self.Base , "Altura: ",self.Altura)
```

```python
class Piso_rodape(Retagulo):
    def calc_piso(self):
         #valor do lado de um piso em cm
        try:
            self.pisosB = float(input("qual tamanho da base em cm? "))
            self.pisosA = float(input("qual tamanho da altura em cm? "))
            a = self.Base / self.pisosB
            b = self.Altura / self.pisosA
            c = a * b
            print(c)
        except ValueError:
            print("coloque valores numericos")
        except ZeroDivisionError:
            print("o piso nao pode ter tamanho 0")
    
    def calc_rodape(self):
        try:
            self.rodape = float(input("qual tamanho da base em cm? "))
            a = self.Base / self.rodape * 2
            b = self.Altura / self.rodape * 2
            c = a + b
            print(c)
        except ValueError:
            print("coloque valores numericos")
        except ZeroDivisionError:
            print("o piso nao pode ter tamanho 0")  
```
